```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hw2.1.ipynb

url = 'https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8'
tables = pd.read_html(url, header=0)[12]
#Замініть у таблиці значення "—" на значення NaN
tables.replace("—", np.nan, inplace=True)
print(tables)

plt.figure(figsize=(10, 6))
plt.bar(tables['Регіон'], tables['2019'].astype(float), color='skyblue')
plt.xlabel('Регіон')
plt.ylabel('Рівень народжуваності у 2019 році')
plt.title('Рівень народжуваності у 2019 році за регіонами')
plt.xticks(rotation=90)  # обертання підписів на осі X для зручності
plt.show()


region_data = tables[tables['Регіон'] == 'Крим']
plt.figure(figsize=(8, 5))
plt.plot(region_data.columns[1:], region_data.iloc[:, 1:].values.flatten(), marker='o')
# plt.bar(region_data['Регіон'], region_data['2019'])
plt.xlabel('Регіон')
plt.ylabel('Рівень народжуваності')
plt.title('Рівень народжуваності для регіону "Крим"')
plt.show()

region_data_ki = tables[tables['Регіон'] == 'Київська']
region_data_c = tables[tables['Регіон'] == 'Житомирська']
plt.figure(figsize=(10,5))
fig, axs = plt.subplots()
axs.plot(region_data_ki.columns[1:], region_data_ki.iloc[:, 1:].values.flatten(), marker='o', label="Київська")
axs.plot(region_data_c.columns[1:], region_data_c.iloc[:, 1:].values.flatten(), marker='o', label="Житомирська")
plt.xlabel('Регіон')
plt.ylabel('Рівень народжуваності')
plt.title('Зрівняння рівня нарожуваності "Київської" і "Житомирської" області')
plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

# Hw2.3.ipynb
df = pd.read_csv('bestsellers with categories.csv')
sns.scatterplot(x='Price', y ='User Rating', data=df)
plt.title(f'Порівняння ціни на книгу до рейтингу яку вона отримала')
plt.xlabel('Ціна')
plt.ylabel('Рейтинг')
plt.xticks(rotation=90)
plt.show()


sns.lineplot(x='Year', y='Price', hue="Genre", data=df)
plt.show()


sns.relplot(x='Year', y='Price',kind='scatter', hue="Genre",col='Genre', data=df)
plt.show()
```
